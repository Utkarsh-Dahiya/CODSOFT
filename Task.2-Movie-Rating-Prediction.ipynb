```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
## Load The Dataset
```

```python
df = pd.read_csv("IMDb Movies India.csv", encoding='latin1')
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
## Filling the Missing Values
```

```python
df['Genre']=df['Genre'].fillna(df['Genre'].mode()[0])
```

```python
df['Director'] = df['Director'].fillna(df['Director'].mode()[0])
```

```python
df['Actor 1'] = df['Actor 1'].fillna(df['Actor 1'].mode()[0])
df['Actor 2'] = df['Actor 2'].fillna(df['Actor 2'].mode()[0])
df['Actor 3'] = df['Actor 3'].fillna(df['Actor 3'].mode()[0])
```

```python
df['Duration'] = df['Duration'].fillna(df['Duration'].mode()[0])
```

```python
df['Duration'] = df['Duration'].str.replace(' min', '')
df['Duration'] = df['Duration'].astype(int)
```

```python
df['Year'].head()
```

```python
df['Year'] = df['Year'].str.replace('(', '', regex=False)
df['Year'] = df['Year'].str.replace(')', '', regex=False)

df['Year'] = df['Year'].astype(int)
```

```python
df['Votes'].head()
```

```python
df['Votes'] = df['Votes'].str.replace(',', '')

df['Votes'] = df['Votes'].astype(int)
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Genre'] = le.fit_transform(df['Genre'])
df['Director'] = le.fit_transform(df['Director'])
df['Actor 1'] = le.fit_transform(df['Actor 1'])
df['Actor 2'] = le.fit_transform(df['Actor 2'])
df['Actor 3'] = le.fit_transform(df['Actor 3'])
```

```python
df.head()
```

```python
X = df.drop('Rating', axis=1)
y = df['Rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)
```

```python
comparison = pd.DataFrame({
    'Actual Rating': y_test,
    'Predicted Rating': y_pred
})

comparison.head(10)
```
